```python
import pandas as pd
doc = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vR1OYsz7cNFWMK9ZXe_6JgTu3kz1ywcVwajRlnKxKyNxot71TfC728ylqnsNuv7dA/pub?gid=641624856&single=true&output=csv',header=[0,1,2])
# como quiero que mi tabla lea las 3 primeras filas como filas propias, utilizo header[0,1,2]
doc
```

```python
doc.info()
```

```python
# primero reemplazamos las comas con puntos para que los datos que corresponden sean leídos como números
doc = doc.replace(',','.',regex=True)
doc
```

```python
# luego borramos las últimas filas que señalan la leyenda de datos, es decir desde la fila 223 hasta la última
doc = doc.drop(doc.index[223:])
doc
```

```python
#borramos las columnas que no poseen información relevante
byeColumns=[('Unnamed: 3_level_0', 'Unnamed: 3_level_1', 'Unnamed: 3_level_2'),
          ('Unnamed: 5_level_0', 'Unnamed: 5_level_1', 'Unnamed: 5_level_2'),
          ('Unnamed: 7_level_0', 'Unnamed: 7_level_1', 'a'),
          ('Unnamed: 9_level_0', 'Unnamed: 9_level_1', 'a'),
          ('Unnamed: 11_level_0', 'Unnamed: 11_level_1', 'Unnamed: 11_level_2'),
          ('Unnamed: 13_level_0', 'Unnamed: 13_level_1', 'b')]
doc.drop(columns=byeColumns,inplace=True)
doc
```

```python
# ahora borramos los espacios en las celdas númericas
columnas_numericas = doc.select_dtypes(include=['number']).columns

doc[columnas_numericas] = doc[columnas_numericas].applymap(lambda x: str(x).strip() if pd.notna(x) else x)
doc[columnas_numericas] = doc[columnas_numericas].apply(pd.to_numeric, errors='ignore')

# cambiamos los tipos de datos a número, es decir float64, menos la columna de países
# ahora volvemos a cambiar los tipos de datos a número, o sea float64
columns_to_convert = doc.columns.drop( ('Unnamed: 1_level_0', 'Country', 'Unnamed: 1_level_2') )
doc[columns_to_convert] = doc[columns_to_convert].apply(pd.to_numeric, errors='coerce')
doc.info()
doc
```

```python
# ahora quitamos los ".. y "-" y reemplazarlos por Nan
doc = doc.replace('..',pd.NA)
doc = doc.replace('—',pd.NA)
doc
```

```python
# ahora borramos las filas que están en blanco
byeRows = [200,206,208,216,219,221]
doc = doc.drop(byeRows)
# reseteamos el índice
doc = doc.reset_index(drop=True)
doc
```

```python
# finalmente, creamos una nueva columna de categorias 'Human development level'
doc['Human Development Level'] = ''
doc.loc[1:67, 'Human Development Level'] = 'Very High'
doc.loc[68:117, 'Human Development Level'] = 'High'
doc.loc[118:162, 'Human Development Level'] = 'Medium'
doc.loc[163:195, 'Human Development Level'] = 'Low'
doc.loc[196:, 'Human Development Level'] = 'Other countries or territories'

# y lo recategorizamos
doc['Human Development Level (order)'] = doc['Human Development Level']
mapper1 = {'Very High': 1, 'High': 2, 'Medium': 3, 'Low': 4, 'Other countries or territories': 5}
doc['Human Development Level (order)'].replace(mapper1, inplace=True)

from pandas.api.types import CategoricalDtype
cat_type = CategoricalDtype(categories=[1, 2, 3, 4, 5], ordered=True)

doc['Human Development Level'] = doc['Human Development Level'].astype('category')

doc = doc.drop(columns=['Human Development Level (order)'])
doc
```

```python
# ahora borramos las filas que ya no necesitamos
byeRows2 = [0,67,117,162,195]
doc = doc.drop(byeRows2)
# reseteamos el índice
doc = doc.reset_index(drop=True)
doc
```
